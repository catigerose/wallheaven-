```python
from bs4 import BeautifulSoup
import time
import requests
import os
from pathlib import Path
import numpy as np
```

```python
keyword = "flower"
#names = np.load("names.npy").tolist()
#len(names)
```

```python
domain = "https://wallhaven.cc/search?q="+keyword+"&categories=100&purity=100&atleast=2560x1440&ratios=16x9%2C16x10&sorting=relevance&order=desc&page=12"
ret = requests.get(domain)
ret.encoding = ret.apparent_encoding 
time.sleep(1)
soup = BeautifulSoup(ret.text, 'html.parser') 
pages0 = soup.find('header',class_="thumb-listing-page-header").find('h2').get_text().split('/' )[-1]
pages = int(pages0)
pages
```

```python
names = []
for i in range(pages): 
    ret = requests.get("https://wallhaven.cc/search?q="+keyword+"&categories=110&purity=100&sorting=relevance&order=desc&page="+str(i+1))
    ret.encoding = ret.apparent_encoding 
    #time.sleep(1)
    soup = BeautifulSoup(ret.text, 'html.parser') 
    div_list = soup.find('section',class_="thumb-listing-page").find_all('li')
    if div_list:
        for dic2 in div_list:
            name0 = dic2.find("figure").attrs['data-wallpaper-id']
            names.append(name0)            
    else:
        i = pages
np.save("names.npy",np.array(names))
```

```python
my_file = Path("./wallhaven/"+keyword)
if my_file.exists():
        print("./wallhaven/"+keyword+"：目录已存在")
else:
        os.makedirs("./wallhaven/"+keyword)
```

```python
for i in range(len(names)):
    try:
        name = names.pop()
        print(i,name)
        ret = requests.get('https://wallhaven.cc/w/'+name)
        ret.encoding = ret.apparent_encoding 
        #time.sleep(1)
        soup = BeautifulSoup(ret.text, 'html.parser') 
        src = soup.find("main").find("section").find("img").attrs['src']
        r = requests.get(src)
        with open("./wallhaven/"+keyword+"/"+name+".jpg", 'wb') as f:      
            f.write(r.content)
    except:
            names3 = np.array(names)
            #srcs2 = np.array(srcs)
            np.save("names.npy",names3)
            #np.save("srcs.npy",srcs2)
            print("出现异常已保存进度,本次下载照片数量为：",i+1)
```
